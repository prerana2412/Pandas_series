# Addition using Append

```python
import pandas as pd
df = pd.DataFrame([[2,4],[5,7]],columns=['a','b'])
df2 = pd.DataFrame([[7,8],[6,4]],columns=['a','b'])
```

```python
df   # consider as Dataset_1
```

```python
df2  # Dataset_2
```

```python
df.append(df2,ignore_index=True)  # append me always rows add honge
```

```python
# if i want add dataset one after another
pd.concat([df,df2]) # axis by default 0
```

```python
pd.concat([df,df2],axis=1,ignore_index=True)
```

# what is append and concat????Similarity and differencs betn them?
#append always add rows at the end but concat add rowwise and columnwise also
